```python
#dv translating

for i in tqdm(range(len(col))): # df.shape[0]
    num = col[i]
    if pd.isna(df.loc['en', num]) == False:
        try:
            lang = 'Divehi (Dhivehi, Maldivian'
            value_en = df.loc["en", num]
            response = openai.chat.completions.create(
            model = "gpt-4o",
            messages = [
                    {"role": "system", "content": "You are a translator."},
                    {"role": "user", "content": f'Translate this text to {lang}: {value_en}'}
                    ],
            temperature = 0.2
            )
            translated_text = response.choices[0].message.content
            df.loc['dv', num] = translated_text
        except:
            pass
```
