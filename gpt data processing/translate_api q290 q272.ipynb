```python
df = pd.read_csv('/Users/rusya/Downloads/lang_ethnicity_groupby.csv')
interview_language = pd.read_csv('/Users/rusya/Downloads/lang_interview_iso_encoding - Лист1.csv')
Q290 = pd.read_csv('/Users/rusya/Downloads/questions/Q290.csv') 

Q290 = Q290.set_index('idx')
interview_language = interview_language.set_index('idx')
```

```python
error = 0
lang_error = 0
```

```python
#Q290 translating

for i in tqdm(range(67, df.shape[0])):    
        idx = df.loc[i, "LNGE_ISO"]
        try:
            lang = interview_language.loc[df.loc[i, "LNGE_ISO"], "lang"]
            if idx != 'en':
                if df.loc[i, "Q290"] > 0:
                    try:
                        num = df.loc[i, "Q290"]
                        value_en = Q290.loc["en", str(int(num))]
                        response = openai.chat.completions.create(
                        model = "gpt-4o",
                        messages = [
                                {"role": "system", "content": "You are a translator."},
                                {"role": "user", "content": f'Translate this text to {lang}: {value_en}'}
                                ],
                        temperature = 0.2
                        )              
                        translated_text = response.choices[0].message.content
                        Q290.loc[idx, str(num)] = translated_text
                    except KeyError as e:
                        error += 1
        except KeyError as e:
            lang_error += 1
```

```python
print(error)
print(lang_error)
```
