```python
questions = [
    'Q158', 'Q159', 'Q160', 'Q161', 'Q162',
    'Q177', 'Q178', 'Q179', 'Q180',
    'Q181', 'Q182', 'Q183', 'Q184', 'Q185', 'Q186', 'Q187', 'Q188', 'Q189',
    'Q190', 'Q191', 'Q192', 'Q193', 'Q194', 'Q195'
]

# Для каждой QXXX создаём колонку QXXX_y_gpt, где 1 = совпадают, 0 = нет
for q in questions:
    gpt_col = f"{q}_gpt"         
    new_col = f"{q}_y_gpt"    
    df_clean[new_col] = (df_clean[q] == df_clean[gpt_col]).astype(int)

df_clean
```

```python
# список кодов, которые нужно удалить
codes_to_drop = [1105, 1190, 1940, 2190, 3750, 3900, 4530, 9000, 9040, 9060, 9900]

# удаление строк, в которых S_INTLANGUAGE содержит указанные коды
df_clean = df_clean[~df_clean['S_INTLANGUAGE'].isin(codes_to_drop)]

# проверка
print("Новая размерность после удаления:", df_clean.shape)
df_clean
```

```python
df_gdp = df_extra[['code', 'year', 'ВВП на душу населения (2024)', 'ICT Development Index 2024', 'ICT Development Index 2024, Universal (personal indicators dominating)', 'ICT Development Index 2024, Meaningfull (infrastructure indicators dominating)']].copy()
df_gdp = df_gdp.rename(columns={
    'code': 'country',
    'year': 'A_YEAR',
    'ВВП на душу населения (2024)': 'GDP_per_capita_2024',
    'ICT Development Index 2024': 'ICT Development Index 2024',
    'ICT Development Index 2024, Universal (personal indicators dominating)': 'ICT Development Index 2024, Universal (personal indicators dominating)',
    'ICT Development Index 2024, Meaningfull (infrastructure indicators dominating)': 'ICT Development Index 2024, Meaningfull (infrastructure indicators dominating)'
})

df_clean = df_clean.merge(
    df_gdp,
    how='left',
    on=['country', 'A_YEAR']
)
```

```python
# 2. Выполни объединение (join/merge) по коду языка
df_main = df_clean.merge(
    df_lang_fsi,
    how='left',
    left_on='S_INTLANGUAGE',
    right_index=True
)

df_main
```

```python
df_main.rename(columns={
    'closely related': 'closely_related',
    'similar to eng': 'similar_to_eng',
    'ling and/or cult differ': 'ling_and_or_cult_differ',
    'cignificant ling and/or cult differ': 'cignificant_ling_and_or_cult_differ',
    'exceptionally difficult': 'exceptionally_difficult'
}, inplace=True)

df_main
```

```python
for col in ['closely_related', 'similar_to_eng', 'ling_and_or_cult_differ',
            'cignificant_ling_and_or_cult_differ', 'exceptionally_difficult']:
    print(col, df_main[col].value_counts(dropna=False))
```

```python
categoricals = [
    'closely_related',
    'similar_to_eng',
    'ling_and_or_cult_differ',
    'cignificant_ling_and_or_cult_differ',
    'exceptionally_difficult'
]

for col in categoricals:
    df_main[col] = df_main[col].astype('category')
```

```python
from statsmodels.formula.api import logit

formula = (
    'Q158_y_gpt ~ '
    'C(closely_related) + '
    'C(similar_to_eng) + '
    'C(ling_and_or_cult_differ) + '
    'C(cignificant_ling_and_or_cult_differ) + '
    'C(exceptionally_difficult) + '
    'GDP_per_capita_2024'
)

model = logit(formula, data=df_main.dropna()).fit(method='bfgs', maxiter=1000)
print(model.summary())
```

```python
from statsmodels.formula.api import logit

model = logit('Q158_y_gpt ~ C(exceptionally_difficult)', data=df_main.dropna())
result = model.fit(method='bfgs', maxiter=1000)
print(result.summary())
```
