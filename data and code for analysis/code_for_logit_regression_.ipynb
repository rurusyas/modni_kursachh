## Великая и могучая logit регрессия

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score, confusion_matrix
```

```python
# 1) Оставляем только нужные колонки (предикторы + целевой + intercept не нужен явно)
columns_for_model = [
    'Q158_y_gpt',  # целевая
    'X003R2',      # AGE
    'Q260',        # GENDER
    'Q275',        # EDUCATION
    'Q288R',       # INCOME
    'Q281',        # OCCUPATIONAL_GROUP
    'Q287',        # SOCIAL_CLASS
    'Q273',        # MARITAL_STATUS
    'Q274',        # HOW_MANY_CHILDREN
    'Q289',         # RELIGIOUS
    'GDP_per_capita_this_year'
]
df_model = df_clean[columns_for_model].copy()

# 2) Dropna — чтобы не было пустых ячеек:
df_model = df_model.dropna()

# 3) Проверим размерность:
print("Размерность df_model:", df_model.shape)
print(df_model.head())
```

```python
from statsmodels.formula.api import logit

formula = (
    'Q158_y_gpt ~ '
    'C(X003R2) +  '
    'C(Q260) + '        # Gender (1 vs 2)
    'Q275 + '           # Education (0..8)
    'C(Q288R) + '       # Income (1,2,3)
    'C(Q281) + '        # Occupational Group (0..11)
    'C(Q287) + '        # Social Class (1..5)
    'C(Q273) + '        # Marital Status (1..6)
    'Q274 + '           # How Many Children (0..3)
    'C(Q289) + '           # Religious (0..9)
    'GDP_per_capita_this_year'

)
print("Формула модели:", formula)
```

```python
# 1) Подготовим саму модель:
model = logit(formula, data=df_model)

# 2) Обучаем методом BFGS (maxiter = 1000, чтобы успело сойтись):
results = model.fit(method='bfgs', maxiter=1000)

# 3) Смотрим краткий вывод:
print(results.summary())
```
