```python
def recode_education(val):
    if val in [0, 1]:
        return 0
    elif val in [2, 3]:
        return 1
    elif val in [4, 5]:
        return 2
    elif val in [6, 7, 8]:
        return 3
    else:
        return None 
    
df_clean['Q275'] = df_clean['Q275'].apply(recode_education)
```

```python
import statsmodels.api as sm

y = df_clean['Q177_gpt']
X = df_clean[['X003R2', 'Q260', 'Q275', 'Q288R', 'Q287', 'Q273', 'Q274', 'Q289']]

X_clean = sm.add_constant(X)
model = sm.OLS(y, X_clean).fit()

print(model.summary())
```
